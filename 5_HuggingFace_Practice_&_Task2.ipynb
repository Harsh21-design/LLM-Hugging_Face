```python
import telebot
from gtts import gTTS

# create a bot object
token = "you telegram api token"
bot = telebot.TeleBot(token)
```

```python
# voice generation from text message func
@bot.message_handler()
def text_to_voice(message):
  text = message.text
  language = "en"
  text_to_speech = gTTS(text=text, lang = language)
  text_to_speech.save("gen_voice.mp3")
  voice = open("gen_voice.mp3", "rb")

  bot.send_voice(message.chat.id, voice)
```

```python
# to run the bot
print("Bot is running")
bot.polling()
```

# Task 2 - Generate voice from text


## Using Hugging Face(Suno/bark-small TTS)

```python
from transformers import pipeline

text = " Hello Harsh! This is me Friday Your AI Assistant [laughs]  How can i help you today"
model_id = "suno/bark-small"

tts = pipeline("text-to-audio",
               model=model_id,
               device="cuda")

speech = tts(text)
```

```python
print(speech)
```

```python
# playing generated speech/audio
from IPython.display import Audio
Audio(speech["audio"], rate=speech["sampling_rate"])
```
